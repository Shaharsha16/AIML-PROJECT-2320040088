```python
from collections import deque

def BFS(graph, Initial_node, goal_node):
    visited = set()
    queue = deque([Initial_node])

    while queue:
        node = queue.popleft()
        if node not in visited:
            visited.add(node)
            print(f"Visited nodes: {node}")
            if node == goal_node:
                print(f"Found node {goal_node} using BFS searching technique!")
                return True
            queue.extend(graph[node])
    print(f"Node {goal_node} not found in BFS search technique!")
    return False
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I', 'J', 'K'],
    'E': [],
    'F': ['L', 'M'],
    'G': ['N'],
    'H': ['O', 'P'],
    'I': ['Q', 'R'],
    'J': ['S', 'T'],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
    'O': [],
    'P': [],
    'Q': [],
    'R': [],
    'S': [],
    'T': []
}
BFS(graph, 'A', 'P')
```

```python
def DFS(graph, Initial_node, goal_node):
    visited = set()
    stack = [Initial_node]

    while stack:
        node = stack.pop()
        if node not in visited:
            visited.add(node)
            print(f"Visited node: {node}")
            if node == goal_node:
                print(f"Found node {goal_node} using DFS searching technique!")
                return True
            stack.extend(reversed(graph[node]))
    print(f"Node {goal_node} not found in DFS search technique!")
    return False
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I', 'J', 'K'],
    'E': [],
    'F': ['L', 'M'],
    'G': ['N'],
    'H': ['O', 'P'],
    'I': ['Q', 'R'],
    'J': ['S', 'T'],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
    'O': [],
    'P': [],
    'Q': [],
    'R': [],
    'S': [],
    'T': []
}
DFS(graph, 'A', 'P')
```

```python
def DLS(graph, node, goal_node, depth, visited):
    if depth == 0:
        if node == goal_node:
            print(f"Found node {goal_node} using DLS with depth {depth}!")
            return True
        return False

    if node not in visited:
        visited.add(node)
        print(f"Visited node: {node} at depth {depth}")

        for neighbor in graph[node]:
            if DLS(graph, neighbor, goal_node, depth - 1, visited):
                return True

    return False

def IDDFS(graph, Initial_node, goal_node, max_depth):
    for depth in range(max_depth):
        visited = set()
        print(f"\nSearching at depth {depth}:")
        if DLS(graph, Initial_node, goal_node, depth, visited):
            return True
    print(f"Node {goal_node} not found within depth limit of {max_depth}.")
    return False
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I', 'J', 'K'],
    'E': [],
    'F': ['L', 'M'],
    'G': ['N'],
    'H': ['O', 'P'],
    'I': ['Q', 'R'],
    'J': ['S', 'T'],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
    'O': [],
    'P': [],
    'Q': [],
    'R': [],
    'S': [],
    'T': []
}

IDDFS(graph, 'A', 'P', 5)
```

```python
import heapq

def heuristic(node, goal_node, h_values):
    """
    This function returns the heuristic value for a node.
    """
    return h_values.get(node, float('inf'))

def BestFirstSearch(graph, Initial_node, goal_node, h_values):
    visited = set()
    priority_queue = []

    heapq.heappush(priority_queue, (heuristic(Initial_node, goal_node, h_values), Initial_node))

    while priority_queue:
        current_heuristic, node = heapq.heappop(priority_queue)

        if node not in visited:
            visited.add(node)
            print(f"Visited node: {node}")

            if node == goal_node:
                print(f"Found node {goal_node} using Best-First Search!")
                return True

            for neighbor in graph[node]:
                if neighbor not in visited:
                    heapq.heappush(priority_queue, (heuristic(neighbor, goal_node, h_values), neighbor))

    print(f"Node {goal_node} not found in Best-First Search!")
    return False
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I', 'J', 'K'],
    'E': [],
    'F': ['L', 'M'],
    'G': ['N'],
    'H': ['O', 'P'],
    'I': ['Q', 'R'],
    'J': ['S', 'T'],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
    'O': [],
    'P': [],
    'Q': [],
    'R': [],
    'S': [],
    'T': []
}

h_values = {
    'A': 10, 'B': 8, 'C': 5, 'D': 7,
    'E': 7, 'F': 6, 'G': 5, 'H': 3,
    'I': 6, 'J': 4, 'K': 7, 'L': 5,
    'M': 4, 'N': 4, 'O': 2, 'P': 0,
    'Q': 4, 'R': 3, 'S': 3, 'T': 3
}

BestFirstSearch(graph, 'A', 'P', h_values)
```

```python
import heapq

def heuristic(node, goal_node, h_values):
    """
    This function returns the heuristic value for a node.
    """
    return h_values.get(node, float('inf'))

def AStarSearch(graph, costs, Initial_node, goal_node, h_values):
    visited = set()
    priority_queue = []
    heapq.heappush(priority_queue, (0 + heuristic(Initial_node, goal_node, h_values), 0, Initial_node))

    while priority_queue:
        current_f, current_g, node = heapq.heappop(priority_queue)

        if node not in visited:
            visited.add(node)
            print(f"Visited node: {node}")

            if node == goal_node:
                print(f"Found node {goal_node} using A* Search!")
                return True

            for neighbor in graph[node]:
                if neighbor not in visited:
                    new_g = current_g + costs[(node, neighbor)]
                    new_f = new_g + heuristic(neighbor, goal_node, h_values)
                    heapq.heappush(priority_queue, (new_f, new_g, neighbor))

    print(f"Node {goal_node} not found in A* Search!")
    return False
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['E', 'F'],
    'C': ['G', 'H'],
    'D': ['I', 'J', 'K'],
    'E': [],
    'F': ['L', 'M'],
    'G': ['N'],
    'H': ['O', 'P'],
    'I': ['Q', 'R'],
    'J': ['S', 'T'],
    'K': [],
    'L': [],
    'M': [],
    'N': [],
    'O': [],
    'P': [],
    'Q': [],
    'R': [],
    'S': [],
    'T': []
}
costs = {
    ('A', 'B'): 2, ('A', 'C'): 7, ('A', 'D'): 8,
    ('B', 'E'): 6, ('B', 'F'): 1,
    ('C', 'G'): 12, ('C', 'H'): 18,
    ('D', 'I'): 2, ('D', 'J'): 1, ('D', 'K'): 12,
    ('F', 'L'): 10, ('F', 'M'): 1,
    ('G', 'N'): 2,
    ('H', 'O'): 3, ('H', 'P'): 4,
    ('I', 'Q'): 5, ('I', 'R'): 6,
    ('J', 'S'): 7, ('J', 'T'): 8,
}
h_values = {
    'A': 10, 'B': 8, 'C': 5, 'D': 7,
    'E': 7, 'F': 6, 'G': 5, 'H': 3,
    'I': 6, 'J': 4, 'K': 7, 'L': 5,
    'M': 4, 'N': 4, 'O': 2, 'P': 0,
    'Q': 4, 'R': 3, 'S': 3, 'T': 3
}

AStarSearch(graph, costs, 'A', 'P', h_values)
```

```python
import math
graph = {
    'A': [('B', 2), ('C', 7), ('D', 8)],
    'B': [('E', 6), ('F', 12), ('G', 12)],
    'C': [('H', 18)],
    'D': [('I', 2), ('J', 2), ('K', 12)],
    'E': [('L', 10)],
    'F': [('M', 1)],
    'G': [('N', 2)],
    'H': [('O', 3), ('P', 4)],
    'I': [('Q', 5)],
    'J': [('R', 6), ('S', 7)],
    'K': [('T', 8)],
}
utility_values = {
    'L': 10, 'M': 1, 'N': 2, 'O': 3, 'P': 4,
    'Q': 5, 'R': 6, 'S': 7, 'T': 8
}

def minimax(node, depth, is_maximizing):
    if node not in graph:
        return utility_values[node]

    if is_maximizing:
        max_eval = -math.inf
        for child, _ in graph[node]:
            eval = minimax(child, depth + 1, False)
            max_eval = max(max_eval, eval)
        return max_eval
    else:
        min_eval = math.inf
        for child, _ in graph[node]:
            eval = minimax(child, depth + 1, True)
            min_eval = min(min_eval, eval)
        return min_eval
optimal_value = minimax('A', 0, True)
print(f"Optimal value (Minimax): {optimal_value}")
```

```python
import math

graph = {
    'A': [('B', 2), ('C', 7), ('D', 8)],
    'B': [('E', 6), ('F', 12), ('G', 12)],
    'C': [('H', 18)],
    'D': [('I', 2), ('J', 2), ('K', 12)],
    'E': [('L', 10)],
    'F': [('M', 1)],
    'G': [('N', 2)],
    'H': [('O', 3), ('P', 4)],
    'I': [('Q', 5)],
    'J': [('R', 6), ('S', 7)],
    'K': [('T', 8)],
}

utility_values = {
    'L': 10, 'M': 1, 'N': 2, 'O': 3, 'P': 4,
    'Q': 5, 'R': 6, 'S': 7, 'T': 8
}

def alpha_beta(node, depth, alpha, beta, is_maximizing):
    if node not in graph:
        return utility_values[node]

    if is_maximizing:
        max_eval = -math.inf
        for child, _ in graph[node]:
            eval = alpha_beta(child, depth + 1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break
        return max_eval
    else:
        min_eval = math.inf
        for child, _ in graph[node]:
            eval = alpha_beta(child, depth + 1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break
        return min_eval

optimal_value_ab = alpha_beta('A', 0, -math.inf, math.inf, True)
print(f"Optimal value (Alpha-Beta Pruning): {optimal_value_ab}")
```
